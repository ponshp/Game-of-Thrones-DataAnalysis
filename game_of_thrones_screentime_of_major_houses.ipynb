```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,15)
```

```python
data = pd.read_csv("got-houses-screentime.csv")
print(type(data), data.shape, sep = "\n")
```

```python
data.head(5)
```

```python
grouped_df = data.groupby("house", sort = False).sum()
grouped_df
```

```python
grouped_df.describe()
```

```python
houses = grouped_df.index.tolist()
seasons = list(grouped_df.columns.values)
print(houses, seasons, sep = "\n")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
fig.suptitle("Starks Got the Highest Screen Time, Followed by Lannisters", fontsize=16)

grouped_df.plot(kind = "bar", stacked = True, fontsize = 13, ax =ax)
```

```python
fig, ax = plt.subplots(nrows=4, ncols=2)
fig.suptitle("Starks Got the Highest Screen Time, Followed by Lannisters", fontsize=16)


plt.subplot(421)
grouped_df["season 1"].sort_values(ascending = True).plot(kind = "barh", title = "season 1")

plt.subplot(422)
grouped_df["season 2"].sort_values(ascending = True).plot(kind = "barh", title = "season 2")

plt.subplot(423)
grouped_df["season 3"].sort_values(ascending = True).plot(kind = "barh", title = "season 3")

plt.subplot(424)
grouped_df["season 4"].sort_values(ascending = True).plot(kind = "barh", title = "season 4")

plt.subplot(425)
grouped_df["season 5"].sort_values(ascending = True).plot(kind = "barh", title = "season 5")

plt.subplot(426)
grouped_df["season 6"].sort_values(ascending = True).plot(kind = "barh", title = "season 6")

plt.subplot(427)
grouped_df["season 7"].sort_values(ascending = True).plot(kind = "barh", title = "season 7")

plt.show()
```

```python
fig, ax = plt.subplots(nrows=3, ncols=3)
fig.suptitle("Trends in Screen Time Over 7 Seasons for Prominent Houses", fontsize=16)

plt.subplot(421)
grouped_df.loc["Stark"].plot(kind = 'line', title = "Stark", fontsize = 11, x_compat=True)

plt.subplot(422)
grouped_df.loc["Lannister"].plot(kind = 'line', title = "Lannister", fontsize = 11, x_compat=True)

plt.subplot(423)
grouped_df.loc["Targaryen"].plot(kind = 'line', title = "Targaryen", fontsize = 11, x_compat=True)

plt.subplot(424)
grouped_df.loc["Greyjoy"].plot(kind = 'line', title = "Greyjoy", fontsize = 11, x_compat=True)

plt.subplot(425)
grouped_df.loc["Tyrell"].plot(kind = 'line', title = "Tyrell", fontsize = 11, x_compat=True)

plt.subplot(426)
grouped_df.loc["Baratheon"].plot(kind = 'line', title = "Baratheon", fontsize = 11, x_compat=True)

plt.subplot(427)
grouped_df.loc["Martell"].plot(kind = 'line', title = "Martell", fontsize = 11, x_compat=True)

plt.subplot(428)
grouped_df.loc["Tully"].plot(kind = 'line', title = "Tully", fontsize = 11, x_compat=True)


plt.show()
```

```python
ax = plt.axes()
ax.set_title("Heatmap showing the Variations in Screen Time of Prominent Houses for 7 Seasons\n", fontsize = 16)
sns.heatmap(grouped_df, annot=True, fmt = ".2f",  linewidth = 1, 
            linecolor = 'k', robust = True, cmap = 'Reds', ax = ax)
```
